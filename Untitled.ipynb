```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("paddy.csv")
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
df=df.dropna()
```

```python
df.head()
```

```python
X = df[['YEAR','METEOROLOGICAL DROUGHT','HYDROLOGICAL DROUGHT','AGRICULTURAL DROUGHT','AREA UNDER CULTIVATION']]
```

```python
y=df['YIELD']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
pred = lm.predict(X_test)
```

```python
print(pred)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(y_test,pred)
plt.xlabel('Y test')
plt.ylabel('Predected y')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("maize.csv")
```

```python
df.head()
```

```python
df=df.dropna()
scaler = StandardScaler()
scaler.fit(df)
```

```python
X = df[['YEAR','METEOROLOGICAL DROUGHT','HYDROLOGICAL DROUGHT','AGRICULTURAL DROUGHT','AREA UNDER CULTIVATION']]
y = df['YIELD']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
pred = lm.predict(X_test)
```

```python
print(pred)
```

```python
plt.scatter(y_test,pred)
plt.xlabel('Y test')
plt.ylabel('Predected y')
```
